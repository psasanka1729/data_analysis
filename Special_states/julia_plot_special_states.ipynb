```julia
using Plots
using DelimitedFiles
using ColorSchemes
#using CSV
using LaTeXStrings
#using PyPlot
```

```julia
file = raw"8_special_states_data_0_256_1000_1256_2000_2256_3000_3256_4000_4256.txt"
M = readdlm(file)
Special_eigenvalues_1 = M[:,1]; # Special_eigenvalues_1
Special_eigenvalues_2 = M[:,2]; # Special_eigenvalues_2
min_energy1 = minimum(Special_eigenvalues_1)
max_energy1 = maximum(Special_eigenvalues_1)
min_energy2 = minimum(Special_eigenvalues_2)
max_energy2 = maximum(Special_eigenvalues_2)
```

## determinent and trace

```julia
trace_of_eigenvalues = []
determinent_of_eigenvalues = []
for i = 1:length(Special_eigenvalues_1)
    push!(trace_of_eigenvalues,Special_eigenvalues_1[i]+Special_eigenvalues_2[i])
    push!(determinent_of_eigenvalues,Special_eigenvalues_1[i]*Special_eigenvalues_2[i])
end
```

```julia
reduced_trace_of_eigenvalues = trace_of_eigenvalues/mean(trace_of_eigenvalues)
histogram(reduced_trace_of_eigenvalues,
    bins = range(0,16, step = 1),
    #xlims=(-0.3,5),
    #ylims=(0.1,0.7),
    xlabel = "trace",
    ylabel = "",
    yguidefontrotation=-90,
    right_margin = 6Plots.mm,
    left_margin = 10Plots.mm,
    color = "snow4",
    label = "L=8",
    title = "Histogram of trace of the special states",
    dpi=500,
    font="CMU Serif",
    titlefontsize=13,
    guidefontsize=13,
    tickfontsize=13,
    legendfontsize=13,
    normalize=true,
    framestyle = :box)
```

```julia
reduced_determinent_of_eigenvalues = determinent_of_eigenvalues/mean(determinent_of_eigenvalues)
histogram(reduced_determinent_of_eigenvalues,
    bins = range(0,16, step = 1),
    #xlims=(-0.3,5),
    #ylims=(0.1,0.7),
    xlabel = "eigenvalue",
    ylabel = "",
    yguidefontrotation=-90,
    right_margin = 6Plots.mm,
    left_margin = 10Plots.mm,
    color = "snow4",
    label = "L=8",
    title = "Histogram of determinent of the special states",
    dpi=500,
    font="CMU Serif",
    titlefontsize=13,
    guidefontsize=13,
    tickfontsize=13,
    legendfontsize=13,
    normalize=true,
    framestyle = :box)
```

## Surmise of Wigner

```julia
S = abs.(Special_eigenvalues_1-Special_eigenvalues_2)
using Statistics
s_w = S/mean(S);
```

```julia
histogram(s_w,
    bins = range(0,16, step = 0.05),
    xlims=(-0.3,5),
    #ylims=(0.1,0.7),
    xlabel = "Energy (E)",
    ylabel = "P(E)",
    yguidefontrotation=-90,
    right_margin = 6Plots.mm,
    left_margin = 10Plots.mm,
    color = "snow4",
    label = "L=8",
    title = "Histogram of eigenvalue spacing of the special states",
    dpi=500,
    font="CMU Serif",
    titlefontsize=13,
    guidefontsize=13,
    tickfontsize=13,
    legendfontsize=13,
    normalize=true,
    framestyle = :box)
#savefig("6_level_spacing_0_256_1000_1256_2000_2256_3000_3256_4000_4256.png")
s = range(0, 5, length=1000)
#plot!(s,(32*s.^2/pi^2).*exp.(-4*s.^2/pi))
plot!(s,pi*(s/2).*exp.(-(pi*s.^2)/4),linewidth=3,lc=:red3,label = L" p(s) = \frac{\pi s}{2} e^{-\pi s^{2}/4}")
```

```julia
##### reduced_eigenvalue_1 = Special_eigenvalues_1/mean(Special_eigenvalues_1)
histogram(Special_eigenvalues_1,
    bins = range(min_energy1,max_energy1, step = 1),
    xlims=(min_energy1,max_energy1),
    #ylims=(0.1,0.7),
    xlabel = "Energy",
    color = "blue1",
    label = "L=8",
    title = "Histogram of eigenvalue 1",
    dpi=500,
    font="CMU Serif",
    titlefontsize=13,
    guidefontsize=13,
    tickfontsize=13,
    legendfontsize=13,
    normalize=true,
    framestyle = :box)
amplitude = 0.072
Gauss_mean = -1.72
standard_dev = 5.51
Gaussian(x) = amplitude*exp.(-(x.-Gauss_mean).^2/(2*standard_dev^2))
x = range(-20, 20, length=1000)
plot!(x,Gaussian(x),linewidth=3,lc=:red3,label = L" p(s) = 0.072 e^{-(x+1.72)^{2}/60.94}")
#savefig("8_eigenvalue_1_0_256_1000_1256_2000_2256_3000_3256_4000_4256.png")
```

```julia
#reduced_eigenvalue_2 = Special_eigenvalues_2/mean(Special_eigenvalues_2)
histogram(Special_eigenvalues_2,
    bins = range(min_energy2,max_energy2, step = 1),
    xlims=(min_energy2,max_energy2),
    #ylims=(0.1,0.7),
    xlabel = "Energy",
    color = "red",
    label = "L=8",
    title = "Histogram of eigenvalue 2",
    dpi=500,
    font="CMU Serif",
    titlefontsize=13,
    guidefontsize=13,
    tickfontsize=13,
    legendfontsize=13,
    normalize=true,
    framestyle = :box)
amplitude = 0.073
Gauss_mean = 1.47
standard_dev = 5.43
Gaussian(x) = amplitude*exp.(-(x.-Gauss_mean).^2/(2*standard_dev^2))
x = range(-20, 20, length=1000)
plot!(x,Gaussian(x),linewidth=3,lc=:blue,label = L" p(s) = 0.073 e^{-(x-1.47)^{2}/58.97}")
#savefig("8_eigenvalue_2_0_256_1000_1256_2000_2256_3000_3256_4000_4256.png")
```
