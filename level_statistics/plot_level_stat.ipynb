```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from math import*
from numpy import linalg as LA
#L = 4
#SEED = 2022
i,r_avg,r_std = np.loadtxt('10_1800_1816_level_stat_plot_data.txt', dtype="complex",delimiter = '\t',unpack=True)
i = i.real
r_avg = r_avg.real
r_std = r_std.real
```
