```julia
using Plots
using DelimitedFiles
using ColorSchemes
using LaTeXStrings
```

```julia
file = raw"8_1800_1809_energy_data.txt"
M = readdlm(file)
energy_index = M[:,1]; # eigenvalue index
energy = M[:,2]; # level stat
#normalized_energy = energy/maximum(energy);
```

```julia
min_energy = minimum(energy)
max_energy = maximum(energy)
```

```julia
file1 = raw"6_1800_1809_energy_data.txt"
M1 = readdlm(file1)
energy_index1 = M1[:,1]; # eigenvalue index
energy1 = M1[:,2]; # level stat
#normalized_energy = energy/maximum(energy);
```

```julia
min_energy1 = minimum(energy1)
max_energy1 = maximum(energy1)
```

```julia
bin = (max_energy-min_energy)/2^8
bin1 = (max_energy1-min_energy1)/2^6
histogram(energy,
    bins = range(min_energy,max_energy, step = 1),
    xlims=(min_energy,max_energy),
    #ylims=(0.1,0.7),
    xlabel = "Average energy",
    color = "ivory4",
    label = "L=8",
    title = "Histogram of average energy",
    dpi=500,
    font="CMU Serif",
    titlefontsize=13,
    guidefontsize=13,
    tickfontsize=13,
    legendfontsize=13,
    normalize=true)

histogram!(energy1,
    bins = range(min_energy1,max_energy1, step = 1),
    xlims=(min_energy,max_energy),
    #ylims=(0.1,0.7),
    xlabel = "Average energy",
    color = "slateblue3",
    label = "L=6",
    title = "Histogram of average energy",
    dpi=500,
    font="CMU Serif",
    titlefontsize=13,
    guidefontsize=13,
    tickfontsize=13,
    legendfontsize=13,
    normalize=true)
#savefig("6_average_energy_histogram_1800_1809.png")
```
